# Exploratory Data Analysis (EDA)

## Load and Explore the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("../data/creditcard.csv")

# Display basic information about the dataset
print(df.info())
print(df.describe())

# Display first few rows
df.head()

# Check for null values
print(df.isnull().sum())
```

## Check Class Distribution

```python
# Count fraud vs non-fraud transactions
fraud_count = df['Class'].value_counts()

# Plot the distribution
plt.figure(figsize=(6,4))
sns.barplot(x=fraud_count.index, y=fraud_count.values, palette=["blue", "red"])
plt.xticks([0, 1], ['Non-Fraud (0)', 'Fraud (1)'])
plt.ylabel("Count")
plt.title("Class Distribution")
plt.show()
```

## Check Correlations

```python
# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

## Visualize Transaction Amounts

```python
# Plot fraud vs non-fraud transaction amounts
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Class'], y=df['Amount'], palette=["blue", "red"])
plt.ylim(0, 300)  # Limit to remove extreme outliers
plt.title("Transaction Amounts by Class")
plt.show()
```
